```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
from cProfile import label

x = [-1, -0.7777, -0.5555, -0.3333, -0.1111, 0.1111, 
0.3333, 0.5555, 0.7777, 1]
y = [-1.1395, -0.5717, -0.2169, 0.5423, 0.4940,
1.1497, 1.6422, 2.1749, 2.6477,2.9568]
plt.plot(x, y, 'o', label = "Dados Originais")
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.show()
```

```python
x,y = np.array(x).reshape(-1,1), np.array(y).reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(x,y)
print('A estimado', reg.coef_.ravel()[0])
print('B estimado', reg.intercept_[0])
```

```python
y_pred = reg.predict(x)
score = reg.score(x,y)
print('score', score)
```

```python
plt.plot(x, y, 'o', label = "Dados Originais")
plt.plot(x, y_pred, label = "Regrassão Linear")
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.title('Regrassão Linear no scikit-learn')
plt.grid()
plt.show()
```
